```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = "wikitable sortable ")

```

```python
table = soup.find_all('table')[0]
```

```python
 print(table)
```

```python
world_titles = table.findAll('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    print(individual_row_data)

    length = len(df)
    df.loc[length] = individual_row_data

```

```python
 df.head()
```

```python

import altair as alt

# Create a bar chart of name against revenue
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Name', sort='-y'),  # Sort bars by revenue in descending order
    y='Revenue (USD millions)',
    tooltip=['Name', 'Revenue (USD millions)']  # Add tooltip for more info
).properties(
    width=800,
    height=500,
    title='Company Revenue'
)

# Rotate x-axis labels for better readability
chart = chart.configure_axisX(labelAngle=-45)

chart
```
